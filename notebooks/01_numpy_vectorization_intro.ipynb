# Week 01 — NumPy Vectorization Intro

## Question

Why do we use NumPy arrays instead of plain Python loops in scientific computing and machine learning?

## Goal

In this notebook, I will compare a plain Python loop with a NumPy vectorized operation and explain why vectorization is useful for ML workflows.

```python
import numpy as np 
import time 
import matplotlib.pyplot as plt
```

## Creating data

We create a large array of numbers. The task is simple: multiply every number by 2.

This is not a machine learning model yet. It is a minimal example of array-based computation.

```python
n = 1_000_000

python_list = list(range(n)) 
numpy_array = np.arange(n)

print(type(python_list)) 
print(type(numpy_array)) 
print(numpy_array.shape)
```

## Plain Python loop 

Here we multiply every element manually using a loop.

```python
start = time.time()

python_result = []

for x in python_list:
    python_result.append(x * 2)

python_time = time.time() - start

print(f"Python loop time: {python_time :.6f} seconds")
```

## NumPy vectorized operation 

Here we ask NumPy to multiply the whole array at once.

```python
start = time.time()

numpy_result = numpy_array * 2

numpy_time = time. time() - start

print(f"NumPy vectorized time: {numpy_time: .6f} seconds")
```

## Checking that both results are equivalent 

The two methods should produce the same result.

```python
python_result_as_array = np.array(python_result) 
are_equal = np.array_equal(python_result_as_array, numpy_result) 
print("Results are equal:", are_equal)
```

## Visual comparison 

Now we compare execution time.

```python
methods = ["Python loop", "NumPy vectorized"]
times = [python_time, numpy_time]

plt.figure(figsize=(6, 4))
plt.bar(methods, times)
plt.ylabel("Time in seconds")
plt.title("Python loop vs NumPy vectorized operation")
plt.show()
```

## Conclusion

The NumPy version is usually faster because the operation is performed at the array level instead of manually looping through Python objects.

This matters for machine learning because models usually process many numbers at once: features, batches, weights, predictions, losses, and gradients.

In future weeks, the same idea will appear in linear regression, matrix multiplication, neural networks, and gradient-based optimization.
